```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
import seaborn as sns

def compute_classification_scores(y_sentiment, y_returns, title, asset):
    accuracy = accuracy_score(y_sentiment, y_returns)
    precision = precision_score(y_sentiment, y_returns, average='macro')
    recall = recall_score(y_sentiment, y_returns, average='macro')
    f1 = f1_score(y_sentiment, y_returns, average='macro')
    
    cm = confusion_matrix(y_sentiment, y_returns)
    # Plot representation {CM -> Sentiment_Class} -----> {0 -> -1} and {1 -> 1} || {0 -> -1} and {1 -> 0} and {2 -> 1}
    plt.figure(figsize=(10, 4))
    if cm.shape[0]==2:
        yticklabels = [-1, 1]
        xticklabels = [-1, 1]
    else:
        yticklabels = [-1, 0, 1]
        xticklabels = [-1, 0, 1]
        
    sns.heatmap(cm, yticklabels=yticklabels, xticklabels=xticklabels, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Returns')
    plt.ylabel('Sentimment')
    plt.title(title)
    plt.savefig(f'Sentiment_OLS_Results/{asset}_ConfusionMatrix.png', transparent=True)
    plt.show()
    
    print(f'Accuracy: {accuracy:.4f}')
    print(f'Precision: {precision:.4f}')
    print(f'Recall: {recall:.4f}')
    print(f'F1 Score: {f1:.4f}')
    
    return accuracy, precision, recall, f1, cm

def apply_format(df):
    format_df = df.iloc[:,1:].applymap('{:.4f}'.format)
    format_df['Asset'] = df.iloc[:,0]
    return(format_df)
```

```python
assets = ['CBU','^GDAXI', 'SnP500', 'GOOG', 'XLE', '^STOXX', '^N225', '^BSESN', '^FTSE', 'BTC-USD']

metric_dict = {'Asset':[], 'Accuracy':[], 'Precision':[], 'Recall':[], 'F1 Score':[]}
returns_dict = {'Asset':[], 'Negative':[], 'Neutral':[], 'Positive':[]}
lag_dict = {}

print(returns_dict['Asset'])
for asset in assets:
    print(asset)
    df_sentiment = pd.read_csv(f'Sentiments1/{asset}_newsSentiment.csv')
    df_result = pd.read_csv(f'LLaMA3.3_results/{asset}_predictions_full_3.3.csv')
    
    returns_df = pd.DataFrame({})
    returns_df['Date'] = df_result['Date'][1:]
    returns_df['Pred_returns'] = (df_result['Median_Prediction'].diff()/df_result['Median_Prediction'][1:])*100
    returns_df['Actual_returns'] = (df_result['Actual_Price'].diff()/df_result['Actual_Price'][1:])*100
    
    merged_df = pd.merge(df_sentiment, returns_df, how='inner', on='Date')
    merged_df['Class_sentiment'] = merged_df['Sentiment'].apply(lambda x: x/np.abs(x) if x!=0 else 0)
    merged_df['Class_pred_returns'] = merged_df['Pred_returns'].apply(lambda x: 1 if x>=0 else -1)
    merged_df['Class_actual_returns'] = merged_df['Actual_returns'].apply(lambda x: 1 if x>=0 else -1)
    
    returns_list = merged_df.groupby('Class_sentiment')[['Pred_returns', 'Actual_returns']].mean()['Actual_returns'].values
    if len(returns_list)==1:
        returns_list = [np.nan, returns_list[0], np.nan]
    if len(returns_list)==2:
        returns_list = [returns_list[0], np.nan, returns_list[1]]
    print(returns_list)
    returns_dict['Asset'] = returns_dict['Asset'] + [asset]
    returns_dict['Negative'] = returns_dict['Negative'] + [returns_list[0]]
    returns_dict['Neutral'] = returns_dict['Neutral'] + [returns_list[1]]
    returns_dict['Positive'] = returns_dict['Positive'] + [returns_list[2]]
    
    accuracy, precision, recall, f1, _ = compute_classification_scores(merged_df['Class_sentiment'], merged_df['Class_actual_returns'], 'Actual', asset)
    metric_dict['Asset'] = metric_dict['Asset'] + [asset]
    metric_dict['Accuracy'] = metric_dict['Accuracy'] + [accuracy]
    metric_dict['Precision'] = metric_dict['Precision'] + [precision]
    metric_dict['Recall'] = metric_dict['Recall'] + [recall]
    metric_dict['F1 Score'] = metric_dict['F1 Score'] + [f1]
    
    cross_corr = []
    for lag in range(-5,6,1):
        corr_lag = merged_df['Sentiment'].corr(merged_df['Actual_returns'].shift(lag))
        cross_corr.append(corr_lag)
    lag_dict[asset] = cross_corr

avg_sent_returns_df = pd.DataFrame(returns_dict)
metric_df = pd.DataFrame(metric_dict)
corr_dict_df = pd.DataFrame(lag_dict, index=range(-5,6,1)).applymap('{:.4f}'.format)

avg_sent_returns_df = apply_format(avg_sent_returns_df)
metric_df = apply_format(metric_df)

avg_sent_returns_df.to_csv('sentiment_returns.csv')
metric_df.to_csv('metric.csv')
corr_dict_df.to_csv('sentiment_lag_corr.csv')
```

```python
assets = ['CBU','^GDAXI', 'SnP500', 'GOOG', 'XLE', '^STOXX', '^N225', '^BSESN', '^FTSE', 'BTC-USD']

for asset in assets:
    print(asset)
    df_sentiment = pd.read_csv(f'Sentiments1/{asset}_newsSentiment.csv')
    df_result = pd.read_csv(f'LLaMA3.3_results/{asset}_predictions_full_3.3.csv')
    
    returns_df = pd.DataFrame({})
    returns_df['Date'] = df_result['Date'][1:]
    returns_df['Pred_returns'] = (df_result['Median_Prediction'].diff()/df_result['Median_Prediction'][1:])*100
    returns_df['Actual_returns'] = (df_result['Actual_Price'].diff()/df_result['Actual_Price'][1:])*100
    
    merged_df = pd.merge(df_sentiment, returns_df, how='inner', on='Date')
    merged_df['Class_sentiment'] = merged_df['Sentiment'].apply(lambda x: x/np.abs(x) if x!=0 else 0)
    merged_df['Class_pred_returns'] = merged_df['Pred_returns'].apply(lambda x: 1 if x>=0 else -1)
    merged_df['Class_actual_returns'] = merged_df['Actual_returns'].apply(lambda x: 1 if x>=0 else -1)

    x = merged_df['Sentiment']
    y = merged_df['Actual_returns']
    plt.scatter(x, y, s=5, color='blue')
    
    # adding the constant term
    x_const = sm.add_constant(x)
    
    # performing the regression
    # and fitting the model
    result = sm.OLS(y, x_const).fit()
    
    min_x = min(x)
    max_x = max(x)
    # range of values for plotting
    # the regression line
    x_lin_fit = np.arange(-1, 1, 0.01)
    # the substituted equation
    y_lin_fit = result.params['Sentiment'] * x_lin_fit + result.params['const']
    plt.plot(x_lin_fit, y_lin_fit, color='red', linestyle='--')
    plt.xlabel('Sentiment Score')
    plt.ylabel('Percentage Returns')
    sign = '+' if result.params['const']>=0 else '-'
    plt.text(-1, y.max()-0.1*(y.max()-y.min()), f'Y = {result.params['Sentiment']:.2f}X {sign} {abs(result.params['const']):.2f}\nP value = {result.pvalues['Sentiment']:.3f}')
    plt.savefig(f'Sentiment_OLS_Results/{asset}_OLS_LinearFit.png', transparent=True)
    plt.show()
    
    # printing the summary table
    print(result.summary(), '\n\n\n')
```
